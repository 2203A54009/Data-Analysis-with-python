```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = "/content/insurance.csv"
data = pd.read_csv(file_path)
print(data.head())
```

```python
# Vector
ages = data['age']
smokers = data['smoker']
expenses = data['expenses']
print("Vector - Ages:\n", ages.head(), "\n")
print("Vector - Smokers:\n", smokers.head(), "\n")
print("Vector - Expenses:\n", expenses.head(), "\n")
# Factor
smoker_status = data['smoker'].unique()
print("Factor - Smoker Status:\n", smoker_status, "\n")
# List
data_list = [ages, smokers, expenses]
print("List of Data:\n", [type(item) for item in data_list], "\n")
# Matrix
data_matrix = data[['age', 'expenses']].values
print("Matrix - Age and Expenses:\n", data_matrix[:5], "\n")
# DataFrame
print("DataFrame Head:\n", data.head(), "\n")
```

Generate a scatter plot. Customize it by adding a title, axis labels, text annotations, and changing the plot characters (e.g., circles to plus signs).

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='age', y='smoker', marker='P')
plt.title("Scatter Plot: Age vs Smoking")
plt.xlabel("Age")
plt.ylabel("Smoking (Yes/No)")
plt.grid(True)
plt.show()
```

 Filter the data based on a condition and updates the plot to reflect the filtered data.

```python
filtered_data = data[(data['age'] >= 20) & (data['age'] <= 40)]

plt.figure(figsize=(8, 6))
sns.scatterplot(data=filtered_data, x='age', y='smoker', marker='x', color='orange')
plt.title("Scatter Plot: Age (20-40) vs Smoking")
plt.xlabel("Age (20-40)")
plt.ylabel("Smoking (Yes/No)")
plt.grid(True)
plt.show()
```

Plot a scatter plot between age(20 to 40) and smoking

```python
expenses_by_smoker = data.groupby('smoker')['expenses'].mean()
print("\nAverage Expenses:")
print(expenses_by_smoker)

plt.figure(figsize=(8, 6))
expenses_by_smoker.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title("Average Expenses: Smokers vs Non-Smokers")
plt.ylabel("Average Expenses")
plt.xlabel("Smoking Status")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

Find the number of smokers gender wise (plot)

```python
smoker_gender_count = data[data['smoker'] == 'yes'].groupby('sex').size()
print("\nNumber of Smokers by Gender:")
print(smoker_gender_count)

plt.figure(figsize=(8, 6))
smoker_gender_count.plot(kind='bar', color=['lightgreen', 'lightcoral'])
plt.title("Number of Smokers by Gender")
plt.ylabel("Count")
plt.xlabel("Gender")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

Plot a scatter plot between age and expenses

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='age', y='expenses', hue='smoker', style='sex', palette='viridis')
plt.title("Scatter Plot: Age vs Expenses")
plt.xlabel("Age")
plt.ylabel("Expenses")
plt.legend(title="Legend", loc='upper right')
plt.grid(True)
plt.show()
```
